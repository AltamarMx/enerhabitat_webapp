```python
import matplotlib.pyplot as plt
import pandas as pd



def set_plot_style(func):
    def wrapper(*args, **kwargs):
        plt.style.use('ggplot')
        plt.figure(figsize=(10, 6))
        return func(*args, **kwargs)
    return wrapper

@set_plot_style
def plot_data(df):
    df.plot(kind='bar')
    plt.show()

# Uso del decorador
data = {'A': [1, 2, 3, 4], 'B': [5, 6, 7, 8]}
df = pd.DataFrame(data)
plot_data(df)
```
